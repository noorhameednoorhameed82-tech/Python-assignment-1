#Continue statement:
**In Python, the continue statement is used inside loops to skip the current iteration and immediately move to the next iteration of the loop.**

```python
for number in range(1, 11):
    if number % 2 == 0:
        continue
    print(number)
```


##Difference between continue and break:
#continue statement:
*Skips the current iteration and moves to the next iteration.*
#break statement:
*Stops the loop entirely, even if the loop condition is not finished.*

###**Difference in output of two program on basic of break position:**

```python
while True:
    num = 15   # Example value

    if num > 10:
        print("Condition 1 is True (num > 10)")

    elif num > 5:
        print("Condition 2 is True (num > 5)")

    elif num > 0:
        print("Condition 3 is True (num > 0)")

    else:
        print("No condition is True")

    break
```

```python
string = "aabbcc"
count = 3

while True:
    if string[0] == 'a':
        string = string[2:]
    elif string[-1] == 'b':
        string = string[:2]
    else:
        count += 1
        break

print(string)
print(count)
```


      
##Difference between these two programs:
***The main difference between the two programs is the position of the break statement inside the loop.**
 In the first program, the break statement is placed outside the if-elif-else block, which means it executes every time the loop runs. Because of this, the loop runs only once, even if a condition is true. In the second program, the break statement is placed inside the else block. This means the loop will only stop when the else condition is executed. If the first if condition is true, the break statement does not run, and the loop continues for another iteration. Therefore, the key difference is that in the first program the loop always stops after one execution, while in the second program the loop continues until the else condition becomes true.*
