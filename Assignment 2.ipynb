```python
*Python operators:*
=============================================

1. Arithmetic Operators:
 These operators are used to perform mathematical operations like addition, subtraction, multiplication, etc.

a = 10
b = 3

print("Arithmetic Operators:")
print("Addition: a + b =", a + b)       # 10 + 3 = 13
print("Subtraction: a - b =", a - b)    # 10 - 3 = 7
print("Multiplication: a * b =", a * b) # 10 * 3 = 30
print("Division: a / b =", a / b)       # 10 / 3 = 3.333...
print("Floor Division: a // b =", a // b) # 10 // 3 = 3 (integer division)
print("Modulus: a % b =", a % b)        # 10 % 3 = 1 (remainder)
print("Exponent: a ** b =", a ** b)     # 10 ** 3 = 1000
print("\n")
```

```python
2. *Assignment Operators*
 Used to assign values to variables. Can also combine with arithmetic.

x = 5   # simple assignment
print("Assignment Operators:")
x += 3  # x = x + 3
print("x += 3 =>", x) # 8
x -= 2  # x = x - 2
print("x -= 2 =>", x) # 6
x *= 4  # x = x * 4
print("x *= 4 =>", x) # 24
x /= 6  # x = x / 6
print("x /= 6 =>", x) # 4.0
x %= 3  # x = x % 3
print("x %= 3 =>", x) # 1.0
print("\n")
```

```python
 3. *Logical Operators*
 Definition: Used to combine conditional statements: and, or, not

p = True
q = False

print("Logical Operators:")
print("p and q =", p and q)  # True if both are True
print("p or q =", p or q)
```

```python
4.*Membership Operators in Python*

 1. 'in' operator: Checks if a value exists in a sequence (like list, tuple, string, etc.).
 2. 'not in' operator: Checks if a value does NOT exist in a sequence.
Example code:

my_list = [10, 20, 30, 40, 50]

Using 'in'
print(20 in my_list)   # True, because 20 is present in the list
print(60 in my_list)   # False, because 60 is not present in the list

 Using 'not in'
print(20 not in my_list)  # False, because 20 is present in the list
print(60 not in my_list)  # True, because 60 is not present in the list

 Another example with string
my_string = "Hello World"
print("Hello" in my_string)      # True
print("Python" not in my_string) # True
```
